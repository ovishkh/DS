```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visuals
sns.set(style="darkgrid")

# Normal Distribution
def plot_normal_distribution():
    mean = 0     # Mean (μ)
    std_dev = 1  # Standard deviation (σ)
    x = np.linspace(-4, 4, 1000)  # Generate 1000 evenly spaced values
    y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Normal Distribution (μ=0, σ=1)")
    plt.title("Normal Distribution")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# Binomial Distribution
def plot_binomial_distribution():
    n = 10        # Number of trials
    p = 0.5       # Probability of success
    x = np.arange(0, n+1)  # Possible number of successes (0 to n)
    y = np.random.binomial(n, p, 10000)  # Simulate 10,000 experiments

    sns.histplot(y, bins=n+1, stat="probability", kde=False)
    plt.title(f"Binomial Distribution (n={n}, p={p})")
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.show()

# Poisson Distribution
def plot_poisson_distribution():
    lam = 3  # Lambda (average rate of occurrence)
    x = np.arange(0, 15)  # Range of occurrences
    y = np.random.poisson(lam, 10000)  # Simulate 10,000 experiments

    sns.histplot(y, bins=15, stat="probability", kde=False)
    plt.title(f"Poisson Distribution (λ={lam})")
    plt.xlabel("Number of Occurrences")
    plt.ylabel("Probability")
    plt.show()

# Plot each distribution
plot_normal_distribution()
plot_binomial_distribution()
plot_poisson_distribution()
```
